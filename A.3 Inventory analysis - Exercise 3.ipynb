```python
dataframes = [plant_a, plant_b, plant_c]
table_names = ['Plant A.csv', 'Plant B.csv', 'Plant C.csv']
```

```python
for df, name in zip(dataframes, table_names):
    df['Source'] = name
```

```python
df = pd.concat(dataframes, ignore_index=True)
df.head()
```

```python
filtered_ps = df[df['Values'] == 'Projected Stock']
filtered_ps.head()
```

```python
len(filtered_ps)
```

```python
week_col = []

for col in df.columns:
    if col.startswith('W'):
        week_col.append(col)
```

```python
week_col
```

```python
filtered_df = filtered_ps[(filtered_ps[week_col] < 0).any(axis=1)]
```

```python
filtered_df.head()
```

```python
len(filtered_df)
```

*Note: the funal number of rows are 278 - these materials has below zero value for projected stock - during any of the weeks.*

```python
def first_oos(material):
    for week in week_col:
        if material[week] < 0:
            return week
    return None
```

```python
filtered_df = filtered_df.copy() 
```

```python
filtered_df.loc[:, 'OOS week'] = filtered_df.apply(first_oos, axis=1)
```

```python
filtered_df.head()
```

```python
filtered_df = filtered_df.copy() 
```

```python
filtered_df.loc[:, 'No of OOS weeks'] = (filtered_df[week_col] < 0).sum(axis=1)
```
